```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
gdp_data = pd.read_csv("Resources/gdp_data_file.csv")
life_data_01 = pd.read_csv("Resources/life_expectancy_01.csv")
life_data = pd.read_csv("Resources/life_expectancy.csv")
gdp_data.head()
```

```python
pd.set_option('display.float_format', str)
```

```python
gdp_df = gdp_data.loc[:,['Country Name', 'Country Code','2008', '2009', '2010', '2011']]
gdp_df['2011'] = gdp_df['2011'].astype(float)
gdp_df['2010'] = gdp_df['2010'].astype(float)
gdp_df['2009'] = gdp_df['2009'].astype(float)
gdp_df['2008'] = gdp_df['2008'].astype(float)
gdp_df['2011'] = gdp_df.apply(lambda x: "${:,}".format(x['2011']), axis=1)
gdp_df['2010'] = gdp_df.apply(lambda x: "${:,}".format(x['2010']), axis=1)
gdp_df['2009'] = gdp_df.apply(lambda x: "${:,}".format(x['2009']), axis=1)
gdp_df['2008'] = gdp_df.apply(lambda x: "${:,}".format(x['2008']), axis=1)
gdp_df = gdp_df.dropna()
gdp_df.head()
```

```python
gdp_df = gdp_df.rename(columns={'Country Name': 'Country_Name',
                                     'Country Code': 'Country_Code'})
gdp_df.head()
```

```python
life_data.head()
```

```python
life_data = life_data.loc[:,['Country Name', 'Country Code', '2008', '2009', '2010', '2011']]
life_data = life_data.round(decimals=2)
life_data.dropna()
life_data.head()
```

```python
life_data = life_data.rename(columns={'Country Name': 'Country_Name',
                                     'Country Code': 'Country_Code'})
life_data.head()
```

```python
merged_df = pd.merge(gdp_df, life_data, on = "Country_Name")
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df = merged_df.rename(columns={'Country_Name': 'Country_Name',
                                     'Country_Code_x': 'Country_Code',
                                     '2008_x': '2008_GDP_Dollars',
                                     '2009_x': '2009_GDP_Dollars',
                                     '2010_x': '2010_GDP_Dollars',
                                     '2011_x': '2011_GDP_Dollars',
                                     'Country_Code_y': 'Country_Code_2',
                                     '2008_y': '2008_Life_Exp_Yrs',
                                     '2009_y': '2009_Life_Exp_Yrs',
                                     '2010_y': '2010_Life_Exp_Yrs',
                                     '2011_y': '2011_Life_Exp_Yrs'})

del merged_df['Country_Code_2']
merged_df.head()
```

```python
gdp_df.to_csv('Output/gdp_table.csv', index=False, header=True)
life_data.to_csv('Output/life_table.csv', index=False, header=True)
merged_df.to_csv('Output/merged_table.csv', index=False, header=True)
```
